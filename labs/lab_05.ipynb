**Dataset**:

Trabajaremos con el archivo `netflix_titles.csv`, que contiene información sobre los títulos disponibles en la plataforma Netflix hasta el año 2021.

| Variable       | Clase     | Descripción                                                                 |
|----------------|-----------|------------------------------------------------------------------------------|
| show_id        | caracter  | Identificador único del título en el catálogo de Netflix.                   |
| type           | caracter  | Tipo de contenido: 'Movie' o 'TV Show'.                                     |
| title          | caracter  | Título del contenido.                                                       |
| director       | caracter  | Nombre del director (puede ser nulo).                                       |
| cast           | caracter  | Lista de actores principales (puede ser nulo).                              |
| country        | caracter  | País o países donde se produjo el contenido.                                |
| date_added     | fecha     | Fecha en la que el título fue agregado al catálogo de Netflix.              |
| release_year   | entero    | Año de lanzamiento original del título.                                     |
| rating         | caracter  | Clasificación por edad (por ejemplo: 'PG-13', 'TV-MA').                      |
| duration       | caracter  | Duración del contenido (minutos o número de temporadas para series).        |
| listed_in      | caracter  | Categorías o géneros en los que está clasificado el contenido.              |
| description    | caracter  | Breve sinopsis del contenido.                                               |




```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
# Cargar datos
df = pd.read_csv('https://raw.githubusercontent.com/fralfaro/ICS40125/main/docs/labs/data/netflix_titles.csv')
df.head()
```

## 📊 Parte 1: Exploración visual básica

1. **Distribución de tipos de contenido**

   * Realiza un gráfico de barras mostrando la cantidad de películas vs series (`type`).
   * Pregunta guía: ¿Cuál es el tipo de contenido predominante en Netflix? ¿Qué implicancias podría tener para los usuarios?

2. **Histograma de años de lanzamiento**

   * Muestra cómo se distribuyen los títulos según su `release_year`.
   * Pregunta guía: ¿Hay más contenido reciente o antiguo? ¿Notas algún patrón en décadas específicas?

3. **Proporción de clasificaciones por edad (`rating`)**

   * Haz un gráfico circular o de barras.
   * Pregunta guía: ¿Qué clasificación es más común? ¿Crees que Netflix está más enfocado en público adulto, familiar o infantil?



```python
# 1. Distribución de tipos de contenido (Movie vs TV Show)
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='type', hue='type', palette='viridis', legend=False)
plt.title('Distribución de Tipos de Contenido en Netflix')
plt.xlabel('Tipo de Contenido')
plt.ylabel('Número de Títulos')
plt.show()

# Pregunta guía: ¿Cuál es el tipo de contenido predominante en Netflix? ¿Qué implicancias podría tener para los usuarios?
print("Las películas son el tipo de contenido predominante. Esto implica que Netflix podría ser más atractivo para usuarios que buscan películas, aunque la oferta de series también es significativa.")

# 2. Histograma de años de lanzamiento
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='release_year', bins=30, kde=True)
plt.title('Distribución de Títulos por Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Número de Títulos')
plt.show()

# Pregunta guía: ¿Hay más contenido reciente o antiguo? ¿Notas algún patrón en décadas específicas?
print(" Hay una clara tendencia hacia contenido más reciente, con un pico de lanzamientos en los últimos años (aproximadamente desde 2015 en adelante).")

# 3. Proporción de clasificaciones por edad (rating)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='rating', hue='rating', order=df['rating'].value_counts().index, palette='magma', legend=False)
plt.title('Proporción de Clasificaciones por Edad en Netflix')
plt.xlabel('Número de Títulos')
plt.ylabel('Clasificación por Edad')
plt.show()

# Pregunta guía: ¿Qué clasificación es más común? ¿Crees que Netflix está más enfocado en público adulto, familiar o infantil?
print("Las clasificaciones 'TV-MA' y 'TV-14' son las más comunes, lo que sugiere que Netflix tiene un fuerte enfoque en el público adulto y adolescente. También hay una cantidad considerable de contenido para público familiar y niños ('TV-PG', 'TV-G', 'PG', 'G').")
```



## 🎨 Parte 2: Tendencias y evolución en el tiempo

4. **Número de títulos agregados por año**

   * Usa `date_added` (convertido a fecha) y grafica una serie temporal.
   * Pregunta guía: ¿Cuándo creció más el catálogo de Netflix? ¿Qué relación podría tener con la expansión internacional de la plataforma?

5. **Heatmap de lanzamientos por año y mes**

   * Construye un mapa de calor con `year_added` y `month_added`.
   * Pregunta guía: ¿Hay meses con más estrenos que otros? ¿Qué conclusiones puedes sacar sobre la estacionalidad de lanzamientos?

6. **Duración de películas por género**

   * Extrae la duración en minutos y crea un boxplot agrupado por género principal.
   * Pregunta guía: ¿Qué géneros suelen durar más? ¿Qué podrías decir sobre la variabilidad dentro de un mismo género?


```python
# Preprocesamiento para Parte 2
df_processed = df.copy()
df_processed['date_added'] = pd.to_datetime(df_processed['date_added'], format='mixed')
df_processed['year_added'] = df_processed['date_added'].dt.year
df_processed['month_added'] = df_processed['date_added'].dt.month

# 4. Número de títulos agregados por año
titles_per_year = df_processed.groupby('year_added').size().reset_index(name='count')

plt.figure(figsize=(12, 6))
sns.lineplot(data=titles_per_year, x='year_added', y='count')
plt.title('Número de Títulos Agregados a Netflix por Año')
plt.xlabel('Año de Incorporación')
plt.ylabel('Número de Títulos')
plt.grid(True)
plt.show()

# Pregunta guía: ¿Cuándo creció más el catálogo de Netflix? ¿Qué relación podría tener con la expansión internacional de la plataforma?
print("El catálogo de Netflix muestra un crecimiento acelerado a partir de 2015-2016, alcanzando su pico alrededor de 2019-2020. Esto coincide con el período de expansión global de la plataforma, invirtiendo fuertemente en contenido original y la adquisición de licencias para diferentes mercados.")

# 5. Heatmap de lanzamientos por año y mes
monthly_yearly_counts = df_processed.groupby(['year_added', 'month_added']).size().unstack(fill_value=0)

plt.figure(figsize=(14, 8))
sns.heatmap(monthly_yearly_counts, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
plt.title('Heatmap de Títulos Agregados por Año y Mes')
plt.xlabel('Mes de Incorporación')
plt.ylabel('Año de Incorporación')
plt.show()

# Pregunta guía: ¿Hay meses con más estrenos que otros? ¿Qué conclusiones puedes sacar sobre la estacionalidad de lanzamientos?
print("Se observa una estacionalidad en la adición de contenido, con algunos meses mostrando consistentemente más lanzamientos a lo largo de los años. A menudo, los lanzamientos se concentran hacia finales de año o en épocas festivas para atraer a la audiencia. Sin embargo, en el dataset, el mes de enero y algunos meses de verano (julio, agosto) muestran picos, lo que podría indicar estrategias de lanzamiento para períodos vacacionales o inicios de año fiscal.")

# 6. Duración de películas por género
movies_df = df_processed[df_processed['type'] == 'Movie'].copy()
movies_df['duration_minutes'] = movies_df['duration'].str.replace(' min', '').astype(float)

# Para simplificar el género principal, tomaremos el primero de la lista 'listed_in'
movies_df['main_genre'] = movies_df['listed_in'].apply(lambda x: x.split(',')[0].strip() if pd.notnull(x) else 'Unknown')

# Filtrar géneros con pocos datos para mejor visualización
genre_counts = movies_df['main_genre'].value_counts()
main_genres_to_plot = genre_counts[genre_counts > 50].index # Considerar géneros con más de 50 películas
movies_df_filtered_genres = movies_df[movies_df['main_genre'].isin(main_genres_to_plot)]

plt.figure(figsize=(14, 8))
sns.boxplot(data=movies_df_filtered_genres, x='duration_minutes', y='main_genre', hue='main_genre', palette='tab10', legend=False)
plt.title('Distribución de la Duración de Películas por Género Principal')
plt.xlabel('Duración (minutos)')
plt.ylabel('Género Principal')
plt.grid(axis='x')
plt.show()

# Pregunta guía: ¿Qué géneros suelen durar más? ¿Qué podrías decir sobre la variabilidad dentro de un mismo género?
print("Algunos géneros como 'Documentaries' y 'Dramas' muestran una duración promedio más alta y una mayor variabilidad, lo que sugiere que estos géneros pueden tener formatos más flexibles. Otros géneros como 'Children & Family Movies' tienden a tener duraciones más cortas y consistentes. La variabilidad dentro de un mismo género es esperable, ya que diferentes producciones pueden apuntar a distintos públicos o estilos dentro del mismo género.")
```



## 🔥 Parte 3: Comparaciones y relaciones

7. **Top 10 países con más producciones**

   * Usa un gráfico de barras para mostrar los países más frecuentes en la columna `country`.
   * Pregunta guía: ¿Qué países dominan el catálogo? ¿Sorprende la distribución o es esperada?

8. **Películas vs Series según género**

   * Realiza un gráfico de barras apiladas para mostrar la cantidad de títulos por género, separando `Movie` y `TV Show`.
   * Pregunta guía: ¿Existen géneros más asociados a películas o a series?

9. **Relación entre duración y año de lanzamiento**

   * Haz un gráfico de dispersión (scatterplot) con `release_year` en el eje X y `duration` (en minutos) en el eje Y.
   * Pregunta guía: ¿Han cambiado los estándares de duración de las películas con el tiempo? ¿Se ven outliers interesantes?



```python
# 7. Top 10 países con más producciones
# Expandir los países si hay múltiples por título
countries_df = df_processed['country'].dropna().str.split(', ').explode()
top_10_countries = countries_df.value_counts().head(10).reset_index()
top_10_countries.columns = ['country', 'count']

plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_countries, x='count', y='country', hue='country', palette='rocket', legend=False)
plt.title('Top 10 Países con Más Producciones en Netflix')
plt.xlabel('Número de Títulos')
plt.ylabel('País')
plt.grid(axis='x')
plt.show()

# Pregunta guía: ¿Qué países dominan el catálogo? ¿Sorprende la distribución o es esperada?
print("Estados Unidos domina claramente el catálogo, seguido por India y Reino Unido.")
print("Esta distribución es esperada, reflejando la fuerte presencia de producciones angloparlantes ")
print("y la creciente inversión en mercados como India.")

# 8. Películas vs Series según género
# Expandir los géneros si hay múltiples por título
genres_df = df_processed.assign(listed_in=df_processed['listed_in'].str.split(', ')).explode('listed_in')
genre_type_counts = genres_df.groupby(['listed_in', 'type']).size().unstack(fill_value=0)
genre_type_counts = genre_type_counts.sort_values(by='Movie', ascending=False).head(15) # Top 15 géneros por películas

plt.figure(figsize=(14, 8))
genre_type_counts.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Número de Películas vs. Series por Género Principal')
plt.xlabel('Género')
plt.ylabel('Número de Títulos')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tipo de Contenido')
plt.tight_layout()
plt.show()

# Pregunta guía: ¿Existen géneros más asociados a películas o a series?
print("Géneros como 'Dramas', 'Comedies' y 'Action & Adventure' tienen una mayor proporción de películas, ")
print("mientras que géneros como 'International TV Shows' y 'Crime TV Shows' están más asociados a series de televisión.")

# 9. Relación entre duración y año de lanzamiento (solo para películas)
# Reutilizamos movies_df que ya tiene 'duration_minutes' y 'release_year'

plt.figure(figsize=(14, 8))
sns.scatterplot(data=movies_df, x='release_year', y='duration_minutes', alpha=0.6, hue='duration_minutes', size='duration_minutes', sizes=(20, 400), legend=False)
plt.title('Relación entre Duración de Películas y Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Duración (minutos)')
plt.grid(True)
plt.show()

# Pregunta guía: ¿Han cambiado los estándares de duración de las películas con el tiempo? ¿Se ven outliers interesantes?
print("No se observa una tendencia clara en los estándares de duración de las películas a lo largo de los años.")
print("La mayoría de las películas se concentran en duraciones de 80-120 minutos. Se pueden observar algunos outliers,")
print("como películas de muy corta o muy larga duración, que no siguen el patrón general.")
```
